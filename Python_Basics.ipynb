# Python Basics


1. What is Python, and why is it popular?
   - Python is a high level, interpreted programming language.
   It is popularly known for its simplicity and its readability and it is easy to learn and platform independent.

2. What is an interpreter in Python?
   - Python's intepreter translates the code and runs it directly. It is basically a program that reads and executes the Python code line by line.

3. What are pre-defined keywords in Python?
   - These are special reserved wordswith pre-defined meanings and purposes in python which are essential for it's syntax and structure.

   Some keywords are : def, if, else, elif, for, while and many more.

4. Can keywords be used as variable names?
   - No, pre-defined keywords are not meant to be used as variable names or as any identifier.

5. What is mutability in Python?
  - Mutability is an ability to modify objects after being created. It is like altering the state of an object after its creation.

  Mutable objects are : Lists, set and dictionaries.

6. Why are lists mutable, but tuples are immutable?
   - Lists can be modified after its creation. Like addition, removal of editing the elements which are already in the memory block. Whereas, tuples doesn't allow its elements to be altered after its creation. Hence lists are mutable while tuples are immutable.

7. What is the difference between “==” and “is” operators in Python?

  - Both operators are used for comparison. While == operator is used to check whether the value of two operands are equal, on the other hand, is operator checks whether two operands refer to the same object in memory block.

8. What are logical operators in Python?
   - Logical operators are used to combine conditions and evaluate them in boolean expressions, basically true or false.

   Logical operators are 'or', 'and', 'not'.

9. What is type casting in Python?
   - It is a process of converting an object from one data type to another.

   Example : Float to integer or vice versa.

10. What is the difference between implicit and explicit type casting?
   - Implicit type casting is done by the python interpreter automatically. Such as for operations on data types float and integer. Interpreter automatically converts the outcome to a float data type. Whereas explicit type casting is handled by the user manually and does not involve the interpreter's intervention.

11. What is the purpose of conditional statements in Python?
    - Conditional statements are used to make decisions and follow different paths based on specific conditions.

12. How does the elif statement work?
    - Elif condition is used when the statement provided in 'if' condition is false. When the condition of 'if' statement is false, elif can be used to check other conditions. Multiple elif conditions can be implemented. If those elif conditions are false as well, 'else' is used to provide the end statement.

13. What is the difference between for and while loops?
    - 'for' loops are used for iterating over a predetermined times or over a known sequence. 'while' loops are used to iterate based on conditions. Like once the statement mentioned is true, the loop halts and the process ends.

14. Describe a scenario where a while loop is more suitable than a for loop.
    - While loop is preferred over for loop when the exact number of iteration is unknown. For example: A turf management records grass types as they arrive. Grass type should be entered in the database unless 'Complete' is entered as the input data.
